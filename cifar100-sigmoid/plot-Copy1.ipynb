```python
import matplotlib.pyplot as plt

input = open('cifar100-sigmoid.log', 'r')

rangeUpdateTime = [0.0]

for line in input:
    line = line.split()
    if 'Valid' in line:
        rangeUpdateTime.append(float(line[-1]))

plt.figure('frame time')
plt.subplot(211)
plt.plot(rangeUpdateTime, '.r',)
plt.grid(True)
plt.subplot(212)
plt.plot(rangeUpdateTime)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
l_n = 0
a_n = 0
loss_x = []
loss_y = []
acc_x = []
acc_y = []
with open("cifar100-sigmoid.log") as f: 
    for line in f:
        line = line.strip()
        if (len(line.split("Valid:")) == 2) and ('Loss' in line):
            print(line)
            print(line.split("Valid:")[1])
            print(line.split("Valid:")[1].split("Loss")[1])
            print(line.split("Valid:")[1].split("Loss")[1].split(" Prec")[0])
            loss_y.append(float(line.split("Valid:")[1].split("Loss")[1].split(" Prec")[0]))
            loss_x.append(l_n)
            l_n += 20   #display
print(len(loss_x))       
plt.figure(figsize=(8,6))
plt.plot(loss_x,loss_y,'',label="loss")
plt.title('loss')
plt.legend(loc='upper right')
plt.xlabel('iter')
plt.ylabel('')
plt.grid(loss_x)
plt.show()
```
